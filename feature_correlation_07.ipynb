# Heatmap을 통해 연속형 변수들 간 상관관계 분석

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.rc('font', family='Malgun Gothic')
```

```python
df = pd.read_csv('./data/종합데이터.csv', index_col = 0)
```

```python
df.columns
```

```python
# 연속형 변수들만 선정
heatmap_data = df[['자본금액', '매출금액', '영업이익', '당기순이익', '자산총합계금액',
       '부채총합계금액', '보증잔액', '예금평잔', '대출평잔']]
```

```python
# 코드는 구글링했어요....
# feature correlation이 1에 가까울수록 상관성이 높음
# 특정 변수들끼리 상관성이 높다면 그 중 하나를 제거하는 것이 좋음 -> 과적합 방지
# ex) 영업이익과 당기순이익의 correlation은 0.93이므로 모델링 시 당기순이익을 제거

colormap = plt.cm.PuBu
plt.figure(figsize=(10, 8))
plt.title("Correlation of Features", y = 1.05, size = 15)
sns.heatmap(heatmap_data.astype(float).corr(), linewidths = 0.1, vmax = 1.0,
           square = True, cmap = colormap, linecolor = "white", annot = True, annot_kws = {"size" : 16})
```
